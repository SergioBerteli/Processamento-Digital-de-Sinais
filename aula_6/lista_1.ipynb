# Quest√£o 5
## A

```python
from numpy import real, imag, roots
import matplotlib.pyplot as plt

numerador = [1, 0, 0] # z¬≤

denominador = [1, -1/4, 1/2] # z^2 -1/4 z + 1/2


zeros = roots(numerador)
polos = roots(denominador)

print(f"Zeros do sistema: {zeros}")
print(f"Polos do sistema: {polos}")

# plot do gr√°fico

plt.figure()
plt.scatter(real(zeros), imag(zeros), marker='o', color='blue', label='Zeros')
plt.scatter(real(polos), imag(polos), marker='x', color='red', label='Polos')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid(True, which='both', linestyle='--', lw=0.5)
plt.xlabel('Real')
plt.ylabel('Imagin√°rio')
plt.title('Polos e zeros da fun√ß√£o de transfer√™ncia')
plt.legend()
plt.show()
```
